```
# author: Asmaa ~ 2019
# ----------------------
# import libraries
import tensorflow as tf
import numpy as np
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense, Activation
```

## Setting Model Parameters

```
# set the structure of NN

# initialize the model
model = Sequential()

# add a layer with one unit and set the dimension of input 
model.add(Dense(units=1, input_shape=[1]))

# set functional properties and compile the model
# Ex: Stochastic Gradient Decsent, Mean Squared Error  
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```
# create sample dataset
# xs ==> inputs
# ys ==> outputs
# the relation is: y = 2*x - 1
xs = np.array([-1.0,  0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

```
# start training, epochs ==> number of iterations
model.fit(xs, ys, epochs=500)
```

## Test Your Model By New Values

```
# expected value: 19
print(model.predict([10.0]))
```
